```python
import pandas as pd

# Part 1: Explore the Data
```

```python
# Import the data from the CSV file
df = pd.read_csv('client_dataset.csv')
df.head()
```

```python
# View the column names
column_names = df.columns
column_names
```

```python
# Use the describe function to gather some basic statistics and round numbers to 2 decimal places
data_description = df.describe().round(2)
data_description


```

```python
# Answering questions:
# Identify the top three item categories with the most entries
top_categories = df['category'].value_counts().head(3)
top_categories

```

```python
# Identify top three items with the highest total sales
top_items = df.groupby('item_id').sum().sort_values(by='unit_price', ascending=False).head(3)
top_items
```

```python
# For the category with the most entries, identify the subcategory with the most entries
most_entries_category = top_categories.index[0]
subcategory_most_entries = df[df['category'] == most_entries_category]['subcategory'].value_counts().index[0]
subcategory_most_entries
print(f'The subcategory with the most entries in the {most_entries_category} category is {subcategory_most_entries}.')


```

```python
# Identify the top 5 clients with the most entries in the data and store their client IDs in a list
top_clients = df['client_id'].value_counts().head(5).index.tolist()
top_clients



```

```python
# Find out how many total units the client with the most entries ordered and store it in a variable
client_with_most_entries = top_clients[0]
total_units_ordered = df[df['client_id'] == client_with_most_entries]['qty'].sum()
total_units_ordered
# Print out the client with the most entries and how many units they ordered
print(f'The client with the most entries is {client_with_most_entries} and they ordered {total_units_ordered} units.')

```

```python
# Part 2: Transform the Data
# Create a column that calculates the subtotal for each line
df['subtotal'] = df['unit_cost'] * df['qty']
df.head()
```

```python
# Create a column for shipping price based on your criteria
df['shipping_price'] = df['qty'] * 0.05
df.head()
```

```python
# Create a column for the total price including sales tax
df['total_price'] = df['subtotal'] + df['shipping_price'] * 1.075
df.head()
```

```python
# Create a column for the cost of each line
df['line_cost'] = df['unit_cost'] * df['qty']
df.head()
```

```python
# Part 3: Confirm Your Work
# Check if the calculated total prices match the given Order IDs and store the results in a list
# Create a list of the order IDs in the data set and store it in a variable
order_ids = df['order_id'].unique().tolist()
order_ids

df['total_price_check'] = df['total_price'] == df['total_price'].sum()
df.head()



```

```python
# Part 4: Summarize and Analyze
# Calculate total revenue from each of the top 5 clients
#Calculate total costs from each of the top 5 clients
# Calculate total proift from each of the top 5 clients
total_revenue = df.groupby('client_id').sum().sort_values(by='total_price', ascending=False)['total_price'].head(5)
total_revenue = total_revenue.round(2)

total_costs = df.groupby('client_id').sum().sort_values(by='total_price', ascending=False)['line_cost'].head(5)
total_costs = total_costs.round(2)

total_profit = total_revenue - total_costs
total_profit = total_profit.round(2)

#Create a summary table with the total revenue, total costs, and total profit for each of the top 5 clients rounded to 2 decimal places
summary_table = pd.concat([total_revenue, total_costs, total_profit], axis=1)
summary_table.columns = ['total_revenue', 'total_costs', 'total_profit']
summary_table = summary_table.round(2)
summary_table


```

```python
# Create a summary DataFrame
summary_df = pd.DataFrame({
    'total_revenue': total_revenue,
    'total_costs': total_costs,
    'total_profit': total_profit
})
summary_df = summary_df.round(2)
summary_df


```

```python
# Format data and columns for presentation
summary_df = summary_df / 1000000  # Convert to millions of dollars
summary_df = summary_df.rename(columns={
    'Total Units Purchased': 'Total Units Purchased (Millions)',
    'Total Shipping Price': 'Total Shipping Price (Millions)',
    'Total Revenue': 'Total Revenue (Millions)',
    'Total Profit': 'Total Profit (Millions)'
})
summary_df = summary_df.round(2)
summary_df


```
